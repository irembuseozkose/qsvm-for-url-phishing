# QSVM for URL Phishing — venv Kurulum Rehberi

Bu notebook, proje için izole bir Python sanal ortamı (`venv`) oluşturmak, gerekli paketleri kurmak ve notebook’ları doğru kernel ile çalıştırmak için hazırlanmıştır.

## 1. Ön koşullar

Sisteminde şunların kurulu olması gerekir:

- Python 3.10 veya 3.11
- `pip`
- VS Code
- VS Code Python eklentisi
- VS Code Jupyter eklentisi

**Not:** Qiskit ve bazı kuantum kütüphaneleri Python 3.12 ile bazen uyumsuzluk çıkarabildiği için bu proje için Python 3.10 veya 3.11 daha güvenlidir.

## 2. Proje klasörünü açma

Terminalde proje klasörüne gir:

```python
cd "QSVM FOR URL PHISHING"
```

VS Code kullanıyorsan bu klasörü doğrudan **File > Open Folder** ile de açabilirsin.

## 3. venv oluşturma

### Windows

```python
python -m venv .venv
```

Eğer birden fazla Python sürümün varsa:

```python
py -3.11 -m venv .venv
```

### Linux / macOS

```python
python3 -m venv .venv
```

## 4. venv aktivasyonu

### Windows PowerShell

```python
.\.venv\Scripts\Activate.ps1
```

### Windows CMD

```python
.venv\Scripts\activate.bat
```

### Linux / macOS

```python
source .venv/bin/activate
```

Aktivasyondan sonra terminalin başında genelde `(.venv)` görünür.

## 5. pip güncelleme

```python
python -m pip install --upgrade pip setuptools wheel
```
