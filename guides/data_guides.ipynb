## Data Guides
Turkish guides for teammates are written in this notebook.

### Görev Tanımı

1. Literatürde kullanılan veri setlerinin araştırılması ve uygun veri setinin seçimi ve seçildikten sonra data\raw yoluna yüklenmesi
2. Veri setinin özelliklerinin gösterilmesi için 01_exploration.ipynb dosyasının düzenlenmesi
3. Veri ön işleme adımlarının nasıl olması gerektiğinin anlaşılması açısından önceden yapılmış SVM ve QSVM uygulamalarının araştırılması ve karşılaştırılması
4. Yapılan araştırma sonucunda preprocessing.py dosyası üzerinde gerekli ekleme ve çıkarmaların yapılması
5. preprocessing.py dosyasındaki kodların 02_preprocessing.ipynb dosyasına import edilerek kullanılması.
### AÇIKLAMALAR
#### 1. Adım
Veri githuba yüklenmeyecek. Herkes kendi makinesine manuel yükleyecek ve dosya yoluna yerleştirilecek. 
#### 2. Adım
Bu adım proje gidişatında önem taşımadığından sonraki aşamalara bırakılacak.
#### 4. Adım
Araştırmalar sonucunda gerekli görülürse kuantum kısmı için ayrı preprocessing modeli oluşturulacak.
#### 5. Adım
preprocessing.py dosyasındaki kodlar notebook üzerinde çağırılacak. Modüler yapı korunacak. Bu adımda verinin işlendikten sonra data\processed yoluna kaydedilmesi gerekmektedir.
#### Zamanlama
Projenin bu adımı diğer görev bölümlerinin ilerleyebilmesi için erken tamamlanmalıdır. Bu sebeple bu görevi alacak kişinin hızlı çalışması elzemdir. Sonrasında farklı görevler verilebilir veya bir başka görev bölümü paylaşılabilir.
