```python
import pandas as pd

# Elinle oluşturduğun yola göre okuyoruz
df = pd.read_csv('../data/raw/Phishing_Legitimate_full.csv')

# İlk 5 satırı göster
df.head()
```

```python
# Veri seti hakkında genel bilgi (Hangi sütun sayı, hangisi boş gibi)
df.info()

# Sayısal değerlerin özeti (Ortalama, min, max değerler)
df.describe()
```

```python
# id sütununu siliyoruz
df.drop('id', axis=1, inplace=True)

print("Yeni sütun sayısı:", df.shape[1])
```

```python
# Korelasyon matrisini hesapla
corr_matrix = df.corr()

# CLASS_LABEL ile olan ilişkilere bak (mutlak değerce büyük olanlar en önemlileridir)
top_features = corr_matrix['CLASS_LABEL'].sort_values(ascending=False).head(11)
print("Phishing ile en güçlü bağlantısı olan 10 özellik:")
print(top_features)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Sadece seçtiğimiz bu en güçlü 10 özelliği ve hedef etiketi alalım
selected_features = ['FrequentDomainNameMismatch', 'PctNullSelfRedirectHyperlinks', 
                     'InsecureForms', 'NumDots', 'PctExtHyperlinks', 
                     'NumSensitiveWords', 'PathLevel', 'AbnormalExtFormActionR', 
                     'UrlLengthRT', 'HostnameLength']

X = df[selected_features]
y = df['CLASS_LABEL']

# Normalizasyon işlemi
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Kontrol edelim: İlk satır artık 0-1 arasında mı?
print("Normalize edilmiş ilk satır:\n", X_scaled[0])
```

```python
from sklearn.model_selection import train_test_split

# %80 Eğitim, %20 Test olarak bölüyoruz
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print(f"Eğitim seti boyutu: {X_train.shape}")
print(f"Test seti boyutu: {X_test.shape}")
```
